```python
FILE_NAME = "students.txt"

while True:
    print("\n[1] Add [2] Update [3] Search [4] Delete [5] Exit")
    choice = input("menu: ")

    data_list = []
    try:
        with open(FILE_NAME, "r", encoding="utf-8") as f:
            for line in f:
                data_list.append(line.strip().split(","))
    except:
        pass

    if choice == '1':
        name = input("name: ")
        last = input("lastname: ")
        pob = input("place of birth: ")
        gen = input("gender: ")
        tel = input("telephone number of you: ")
        e_name = input("emergency contact person: ")
        e_tel = input("telephone number: ")

        with open(FILE_NAME, "a", encoding="utf-8") as f:
            f.write(f"{name},{last},{pob},{gen},{tel},{e_name},{e_tel}\n")
        print("done!")

    elif choice == '2':
        target = input("type the name to edit: ")
        for row in data_list:
            if row[0] == target:
                row[4] = input(f"new telephone number (old: {row[4]})): ")
                with open(FILE_NAME, "w", encoding="utf-8") as f:
                    for r in data_list:
                        f.write(",".join(r) + "\n")
                print("done!")

    elif choice == '3':
        search = input("find (name/lastname/gender): ").lower()
        for row in data_list:
            if search in row[0].lower() or search in row[1].lower() or search == row[3].lower():
                print(f"find: {' '.join(row)}")
            else:
              print("not found!")

    elif choice == '4':
        target = input("type the name to delete: ")
        data_list = [row for row in data_list if row[0] != target]
        with open(FILE_NAME, "w", encoding="utf-8") as f:
            for r in data_list:
                f.write(",".join(r) + "\n")
        print("done!")

    elif choice == '5':
        break
```
